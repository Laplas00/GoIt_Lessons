```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import plotly.express as px
```

```python
# Load data
data = pd.read_csv('./Housing.csv')

# Extract features and target variable
X_d = data.iloc[:, 1:4]
X_d.columns = ['x1', 'x2', 'x3']
X = np.concatenate([
    np.ones((len(X_d), 1)),
    np.array(X_d.x1).reshape(-1, 1),
    np.array(X_d.x2).reshape(-1, 1),
    np.array(X_d.x3).reshape(-1, 1)],
    axis=1)
y = np.array(data.iloc[:, 0]).reshape(-1, 1)

# Initialize linear regression model
model = LinearRegression()

# Training the model
model.fit(X, y)

# # Making predictions on the training data
y_hat = model.predict(X)

# # Visualize the results
fig1 = px.scatter_3d(x=X[:, 1], y=X[:, 2], z=y.reshape(-1))
fig2 = px.scatter_3d(x=X[:, 1], y=X[:, 2], z=y_hat.reshape(-1))
fig1.show()
fig2.show()

# Making predictions on new data
X_new = np.array([[1, 4700, 2, 1]])
y_prediction = model.predict(X_new)

print('y =', y_prediction)
```
